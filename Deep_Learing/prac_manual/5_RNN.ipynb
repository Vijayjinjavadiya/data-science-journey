## Pre-processing

```python
# converting data to lower case

df['review'] =df["review"].str.lower()

```

```python
# Removing the URLs

import re  # re helps computers understand and manipulate text patterns automatically.

def remove_url (text):
    text = re.sub(r"http\S+","",text) #(pattern , repl, string) \S -> non-white space \s -> white space 
    return text
df['review'] = df['review'].apply(remove_url)
```

```python
df.head
```

```python
# removing punctuations 
def remove_pun_ (text):
    text = re.sub(r"[^A-Za-z0-9\s]","",text) # if value is not in A-Z a-z 0-9 \s its called punctuation
    return text

df['review'] = df['review'].apply(remove_pun_)
```

```python
df.head
```

```python
#remove html tags 

def remove_tags(text):
    text = re.sub(r"<.*?>","",text) # here . mean any character * means infinate char size ? means non greedy manner 
    return text
df['review'] = df['review'].apply(remove_tags)
```

```python
df.head
```

```python
# removing the stopword

import nltk # natural language toolkit is a leading platform for building python programs to work with humun language.

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
    tokens = word_tokenize(text)
    stop_words = stopwords.words("english")

    for word in tokens:
        if word in stop_words:
            text = text.replace(word,"")

    return text

df['review'] = df['review'].apply(remove_stopwords)        
```

```python
df.head
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

def stemming(text):
    ps = PorterStemmer()
    stemmed_words = []

    tokens = word_tokenize(text)

    for token in tokens:
        stemmed_token = ps.stem(token)
        stemmed_words.append(stemmed_token)

    return " ".join(stemmed_words)

df['review'] = df['review'].apply(stemming)
```
